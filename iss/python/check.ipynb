```python
import re

def parse_polynomial(poly_str):
    terms = re.findall(r'([-+]?\d*)x\^(\d+)|([-+]?\d*)x|([-+]?\d+)', poly_str)
    coefficients = []
    exponents = []
    # print(terms)
    for term in terms:
        coefficient, exponent,single, constant = term
        if coefficient:
            ac = int(coefficient)
        elif exponent:# and not coefficient:
            ac = 1
            # print("ji")
        elif constant:
            ac = int(constant)
        elif single:
            if single == '-': ac = -1
            elif single == '+': ac = 1
            else: ac = int(single)
        else:
            ac = 1
        coefficients.append(ac)
        if exponent:
            ae = int(exponent)
        # elif coefficient:
        #     ae = 1
        else:
            # if coefficient:
            #     ae = 1
            # else:
                ae = 0
        # print(ae)
        if single:
            ae = 1
        if not coefficient and not constant and not single and not exponent:
            ae = 1
        exponents.append(ae)
    # print(coefficients)
    # print(exponents)    
    return dict(zip(exponents, coefficients))

def multiply_polynomials(poly1, poly2):
    product = {}
    # print(poly1)
    # print(poly2)
    for exp1, coef1 in poly1.items():
        for exp2, coef2 in poly2.items():
            new_exp = exp1 + exp2
            new_coef = coef1 * coef2
            if new_exp in product:
                product[new_exp] += new_coef
            else:
                product[new_exp] = new_coef
    # print(product)
    return product

def format_polynomial(poly):
    terms = []
    for exp, coef in sorted(poly.items(), reverse=True):
        if coef != 0:
            sign = '+' if coef > 0 and len(terms) > 0 else ''
            term = f"{sign}{coef}x^{exp}" if exp > 0 else f"{sign}{coef}"
            terms.append(term)
    
    return ''.join(terms)

def multiply_and_format_polynomials(poly_str1, poly_str2):
    poly1 = parse_polynomial(poly_str1)
    poly2 = parse_polynomial(poly_str2)
    print(poly1)
    print(type(poly1))
    product = multiply_polynomials(poly1, poly2)
    print(type(product))
    print(product)
    result_str = format_polynomial(product)
    
    return result_str

# Example usage:
a = input("Stringinput: ")
a=a.replace("(","")
a=a.split(")")
poly_str1 = a[0]
poly_str2 = a[1]

result = multiply_and_format_polynomials(poly_str1, poly_str2)
if result:
    print(result)
else: print("0")
```
