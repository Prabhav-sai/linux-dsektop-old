```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
# Image dimensions
height = 100
width = 100

# Generate a simple image with vertical and horizontal lines
# image = np.zeros((height, width))
# image[:, width // 2] = 1  # Vertical line
# image[height // 2, :] = 1  # Horizontal line
image = cv.imread("")

# Compute FFT
fft_result = np.fft.fft2(image)

# Shift zero frequency component to the center
fft_result_shifted = np.fft.fftshift(fft_result)

# Plot the original image
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

# Plot the FFT result
plt.subplot(1, 2, 2)
plt.imshow(np.abs(fft_result_shifted), cmap='gray')
plt.title('FFT Result (Magnitude)')
plt.colorbar()

plt.tight_layout()
plt.show()
```
