```python
import pandas as pd
```

```python
fert_df = pd.read_csv('Punjab 77th round.csv',index_col="Total monthly consumption expenditure (INR)")
lit_df = pd.read_csv('Punjab 77th round.csv',index_col="Total Monthly Income (INR)")
```

```python
print(fert_df.head())
```

```python
print(lit_df.head())
```

```python
merged_df = pd.merge(fert_df,lit_df, on="HHID", how ="inner")#inner is intersection, outer is union, left and right try
```

```python
print(merged_df.head())
```

```python
print(merged_df)
```

```python
total_female_df = merged_df[["Total monthly consumption expenditure (INR)", "Total Monthly Income (INR)"]]
```

```python
print(total_female_df.head())
```

```python
total_female_corr = total_female_df["Total monthly consumption expenditure (INR)"].corr(total_female_df["Total Monthly Income (INR)"])# a corr b and b corr a are equal
print(total_female_corr)
```
