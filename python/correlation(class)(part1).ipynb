```python
import pandas as pd
```

```python
fert_df = pd.read_csv('Punjab 77th round.csv',index_col="Land possessed (hectares)")
lit_df = pd.read_csv('Punjab 77th round.csv',index_col="Monthly income from crop production (INR)")
```

```python
print(fert_df.head())
```

```python
print(lit_df.head())
```

```python
merged_df = pd.merge(fert_df,lit_df, on="HHID", how ="inner")#inner is intersection, outer is union, left and right try
```

```python
print(merged_df.head())
```

```python
print(merged_df)
```

```python
total_female_df = merged_df[["Land possessed (hectares)", "Monthly income from crop production (INR)"]]
```

```python
print(total_female_df.head())
```

```python
total_female_corr = total_female_df["Land possessed (hectares)"].corr(total_female_df["Monthly income from crop production (INR)"])
print(total_female_corr)
```
