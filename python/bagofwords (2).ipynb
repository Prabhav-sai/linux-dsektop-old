```python
import re
from nltk.corpus import stopwords
from collections import defaultdict
from nltk.stem import SnowballStemmer
from nltk.stem import WordNetLemmatizer
```

```python
filename = "BAG OF WORDS.txt"
infile = open(filename,'r')
text = infile.read()
infile.close()
```

```python
print(text[:100])
```

```python
#Tokenise

unrefinedStr = re.sub(r'\—|\'|\`|\"|\||\.|\*|\[|\{|\}|\(|\)|\]|\;|\:|\,|\^|\=|\-|\+|\_|\!|\?|\/|\>|\<|\&|\\|\#|\n', r' ', text) # removing special characters
```

```python
print(unrefinedStr)
```

```python
tokens = unrefinedStr.split()
```

```python
print(len(tokens))
print(tokens[:20])#print(tokens[0:20:1])
```

```python
tokens = [myword.lower() for myword in tokens]
```

```python
print(tokens[:20])
```

```python
import nltk
nltk.download('stopwords')
```

```python
mystopwords = set(stopwords.words("english"))
print(len(mystopwords))
print(mystopwords)
```

```python
#stopwordsdict={}
#for stop in mystopwords:
#    stopwordsdict[stop] = 1
```

```python
Bag=[]
for word in tokens:
    if word not in mystopwords and word.isnumeric()==False and len(word)>2:
        Bag.append(word)
```

```python
print(len(Bag))
```

```python
SnowStem = SnowballStemmer('english')
Lemmatiser = WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
```

```python
Stembag = []
Lembag = []
for word in Bag:
    Stembag.append(SnowStem.stem(word))
    Lembag.append(Lemmatiser.lemmatize(word))
    
```

```python
print(Bag[:20])
print(Stembag[:20])
print(Lembag[:20])
```

```python
mydict = {}
for word in Stembag:
    if word in mydict.keys():
        mydict[word] +=1
    else:
        mydict[word] = 1
```

```python
mydict1 = {}
for word in Lembag:
    if word in mydict1.keys():
        mydict1[word] +=1
    else:
        mydict1[word] = 1
```
