```python
import pandas as pd
```

```python
fert_df = pd.read_csv('Punjab 77th round.csv',index_col="Land owned (hectares)")
lit_df = pd.read_csv('Punjab 77th round.csv',index_col="HH size")
```

```python
merged_df = pd.merge(fert_df,lit_df, on="HHID", how ="inner")#inner is intersection, outer is union, left and right try
```

```python
total_female_df = merged_df[["Land owned (hectares)", "HH size"]]
```
