```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = "1"
req = requests.get(URL)
```

```python
soup = BeautifulSoup(req.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
tableRows = soup.find_all('tr')
```

```python
print(len(tableRows))
```

```python
myData = {
    "Acronym": [],
    "Name": [],
    "Price":[],
    "Change":[],
    "PerChange":[],
    "Volume":[],
    "AvgVol":[],
    "MarketCap":[],
    "PE Ratio":[]
}
```

```python
temp = tableRows[1]
tempTemp = temp.find_all('td')
print(len(tempTemp))
```

```python
for i in range(1,len(tableRows)):
    row = tableRows[i]
    rowValues = row.find_all('td')
    myData["Acronym"].append(rowValues[0].text)
    myData["Name"].append(rowValues[1].text)
    myData["Price"].append(rowValues[2].text)
    myData["Change"].append(rowValues[3].text)
    myData["PerChange"].append(rowValues[4].text)
    myData["Volume"].append(rowValues[5].text)
    myData["AvgVol"].append(rowValues[6].text)
    myData["MarketCap"].append(rowValues[7].text)
    myData["PE Ratio"].append(rowValues[8].text)
```

```python
print(myData["Name"])
```

```python
filename = "yahooInfo.csv"
myRedditDataframe = pd.DataFrame(myData)
myRedditDataframe.to_csv(filename, index=False)
myRedditDataframe
```
