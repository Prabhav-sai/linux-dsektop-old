```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://theprint.in/page/"
kewWords = ["gaza", "israel", "palestine", "hamas"]

MyURL = []

# This loop is for the page number. from 1 to 4.
for i in range(2):

    # before appending: https://theprint.in/page/
    newURL = URL + str(i+1) + "/?s="
    # after appending: https://theprint.in/page/1/?s=

    # This loop is for the keywords.
    for words in kewWords:
        newnewURL = newURL + words
        # after appending https://theprint.in/page/1/?s=myKeyWord
        print(newnewURL)
        
        # Storing all the URLs in a list (array).
        MyURL.append(newnewURL)
```

```python
reqs =[]
for urls in MyURL:
    reqs.append(requests.get(urls))
```

```python
print(len(reqs))

UrlLen = len(reqs)
```

```python
print(reqs[0].content)
```

```python
SecondaryLinks=[]
for i in range(UrlLen):
    page= BeautifulSoup(reqs[i].content,'html5lib')
    Links=page.find_all(class_='td_module_16')
    for link in Links:
        print(link.find('a').get('href'))
        SecondaryLinks.append(link.find('a').get('href'))
```

```python
NDSecondaryLinks = list(set(SecondaryLinks))
```

```python
print(len(NDSecondaryLinks))

NoOfLinks = len(NDSecondaryLinks)
```

```python
print(NDSecondaryLinks[0])
```

```python
FinalArticles = []
for i in range(NoOfLinks-20):
    req = requests.get(NDSecondaryLinks[i])
    page = BeautifulSoup(req.content,'html5lib')

    bodycontent = page.find_all('p')

    pagebody = [body.text for body in bodycontent[1:-10]]
    pagebody = "".join(pagebody)

    #print(pagebody)

    FinalArticles.append(pagebody)
```

```python
print(FinalArticles[2])
```

```python
onetext =""
for article in FinalArticles[:100]:
    onetext = onetext+ "\n" + article
```

```python
print(onetext)
```

```python
with open('Print.txt','w') as Print:
    Print.write(onetext)
```
